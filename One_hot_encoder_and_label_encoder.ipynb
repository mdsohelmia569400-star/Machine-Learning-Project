```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
data = {'name':['kamal','akram','hanif','tamim','suraiya'],
        'age':[12,13,np.nan,15,16],
        'sub':['math','bangla','Islam','english','Social'],
        'marks':[90,80,np.nan,85,np.nan]}
```

```python
df = pd.DataFrame(data)
df
```

```python
imputer = SimpleImputer(strategy='most_frequent')
df['name'] = imputer.fit_transform(df[['name']]).ravel() #add .ravel()
df['sub'] = imputer.fit_transform(df[['sub']]).ravel() #add .ravel()
```

```python
df
```

```python
imputer = SimpleImputer(strategy='mean')
df['age'] = imputer.fit_transform(df[['age']]).ravel() #add .ravel()
df['marks'] = imputer.fit_transform(df[['marks']]).ravel() #add .ravel()
```

```python
df
```

```python
label_encoder = LabelEncoder()
df['name'] = label_encoder.fit_transform(df['name'])
df['sub'] = label_encoder.fit_transform(df['sub'])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoding = OneHotEncoder()
encoder = one_hot_encoding.fit_transform(df[['sub','name']]).toarray()
columns = one_hot_encoding.get_feature_names_out(['sub','name'])
new_df = pd.DataFrame(encoder, columns=columns)
new_df
```

```python
df = pd.concat([df.drop(['name','sub'], axis=1), new_df], axis=1)
```
