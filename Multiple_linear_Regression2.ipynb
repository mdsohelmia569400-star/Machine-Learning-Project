```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv('multiple_regression_dataset.csv')

# Step 2: Define features (X) and target (y)
X = df[['TV_Advertising', 'Radio_Advertising', 'Online_Advertising']]
y = df['Sales']

# Step 3: Create and train the model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict using the trained model
y_pred = model.predict(X)

# Step 5: Print model results
print("✅ Coefficients:", model.coef_)
print("✅ Intercept:", model.intercept_)
print("✅ R-squared:", r2_score(y, y_pred))
print("✅ Mean Squared Error:", mean_squared_error(y, y_pred))

# Step 6: Predict on new input data
new_data = pd.DataFrame({
    'TV_Advertising': [100, 150],
    'Radio_Advertising': [20, 30],
    'Online_Advertising': [60, 90]
})
new_predictions = model.predict(new_data)
print("\n📊 Predicted Sales for New Data:\n", new_predictions)

# Step 7: Visualization - Actual vs Predicted
plt.figure(figsize=(8,6))
plt.scatter(y, y_pred, color='blue', alpha=0.6, label="Predicted vs Actual")
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2, label="Perfect Prediction Line")
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Multiple Linear Regression: Actual vs Predicted Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 7: Create DataFrame showing actual and predicted values
comparison_df = pd.DataFrame({
    'TV_Advertising': X['TV_Advertising'],
    'Radio_Advertising': X['Radio_Advertising'],
    'Online_Advertising': X['Online_Advertising'],
    'Actual_Sales': y,
    'Predicted_Sales': y_pred.round(2)
})

# Display first 10 rows
print(comparison_df.head(10))
```
